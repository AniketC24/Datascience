```python
'''

1. If DataColumn having less than 5% of Misssing Values, We can impute with Mean or Median in case of Numrical columns and imputation with Mode will be for  categorical Column.
2. if DataColumn having missing values between 5% to 15% then random value can  be imputed or better to replace with Median or use KNN imputer.
3. if DataColumn having more than 15% of missing values then we can drop that column.

'''

print(df.isnull().sum() / len(df) *100)
```

```python
# Data Cleaning and Preprocessing
'''
1. Based on Following State, we find that age column is non Normal Distributed, since, Mean =! Median.
2. Calculated Upper Bound Value and Lower Bound Value
Range = df['age'].quantile(0.25),df['age'].quantile(0.75)
'''
print(df['age'].agg(['mean','median']))
print(df['age'].quantile(0.25),df['age'].quantile(0.75))
```

```python
# Imputing age Column
df['age'].fillna(df['age'].median(),inplace = True)
```

```python
# Evaluating Descriptive Stats : Method 02
# EDA : Univariate Analysis


import pandas as pd
from collections import OrderedDict

# Select numerical columns
numerical_col = df.select_dtypes(include='number').columns.tolist()


stats = []

for i in numerical_col:
    numerical_stats = OrderedDict({
        'Feature': i,
        'Count': df[i].count(),
        'Mean': df[i].mean(),
        'Median': df[i].median(),
        'Minimum': df[i].min(),
        'Maximum': df[i].max(),
        'Range': df[i].max() - df[i].min(),
        'Standard Deviation': df[i].std(),
        'Skewness': df[i].skew(),
        'Kurtosis': df[i].kurt(),
        'Q1': df[i].quantile(0.25),
        'Q3': df[i].quantile(0.75),
        'IQR': df[i].quantile(0.75) - df[i].quantile(0.25),
    })
    stats.append(numerical_stats)

# Convert to DataFrame
report = pd.DataFrame(stats)
report
```

```python
'''
Insights
1. Mean != Median : The Datasets in Non Normal Distribution
2. Mesokurtic (kurtisis = 3)
   Leptokurtic (kurtosis > 3)
   Platykurtic (kurtosis < 3)

'''
```
